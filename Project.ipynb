```python
from openpyxl import Workbook
from openpyxl.chart import LineChart, Reference

wb = Workbook()
ws = wb.active
ws.title = "Project"

data = [
    ["Stock Price", "Month"],
    [150, "January"],
    [167, "February"],
    [175, "March"],
    [120, "April"],
    [140, "May"],
    [135, "June"],
    [160, "July"],
    [150, "August"],
    [173, "September"],
    [180, "October"],
    [168, "November"],
    [155, "December"]
]

for row in data:
    ws.append(row)

chart = LineChart()
chart.title = "Stock Price by Month"
chart.x_axis.title = "Month"
chart.y_axis.title = "Stock Price"

data_ref = Reference(ws, min_col=1, min_row=1, max_row=13)
cats_ref = Reference(ws, min_col=2, min_row=2, max_row=13)

chart.add_data(data_ref, titles_from_data=True)
chart.set_categories(cats_ref)

ws.add_chart(chart, "D2")

wb.save("Project.xlsx")

print("Stock Price Excel Sheet Created")
```
